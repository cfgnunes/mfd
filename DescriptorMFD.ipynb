## Example: Describing an image

```python
from descriptor.mfd import MFD
```

```python
# Load a image
image = cv2.imread("images-sample/lena.jpg", cv2.IMREAD_GRAYSCALE)

pf.plot_image_list([image], scale_factor=4)
```

Computing the feature descriptor for this image:

```python
mfd_descriptor = MFD()

descriptor = mfd_descriptor.compute_descriptor(image)

print(descriptor)
```

Plot the descriptor into a bar graph.

```python
dict_names = (
    ("Orientation 1", "b"),
    ("Orientation 2", "g"),
    ("Orientation 3", "r"),
    ("Orientation 4", "c"),
    ("Orientation 5", "m")
)

pf.plot_descriptor(descriptor, dict_names, 15)
```
