## Example: Describing an image

```python
from descriptor.eoh import EOH
```

```python
# Load a image
image = cv2.imread("images-sample/lena.jpg", cv2.IMREAD_GRAYSCALE)

pf.plot_image_list([image], scale_factor=4)
```

Computing the feature descriptor for this image:

```python
eoh_descriptor = EOH()

descriptor = eoh_descriptor.compute_descriptor(image)

print(descriptor)
```

Plot the descriptor into a bar graph.

```python
dict_names = (
    ("Horizontal filter", "b"),
    ("Vertical filter", "g"),
    ("45¬∫ filter", "r"),
    ("135¬∫ filter", "c"),
    ("No orientation filter", "m")
)

pf.plot_descriptor(descriptor, dict_names, 10)
```
